# Descriptive Statistics Problems

## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title
df["title_length"] = df["title"].apply(len)
df.head()
```

```python
# Calculate mean, median, and mode of the 'title_length' column
df.title_length.mean(), df.title_length.median(), df.title_length.mode()[0]
print(f'Mean: {df.title_length.mean()}, Median: {df.title_length.median()}, Mode: {df.title_length.mode()[0]}')
```

```python
# Calculate the range, variance, and standard deviation of 'title_length'
title_length_range = df.title_length.max() - df.title_length.min()
title_length_variance = df.title_length.var()
title_length_std = df.title_length.std()
print(
    f"Range: {title_length_range}, Variance: {title_length_variance}, Standard Deviation: {title_length_std}"
)
```

```python
# Calculate skewness and kurtosis of the 'title_length' column
title_length_skewness = skew(df.title_length)
title_length_kurtosis = kurtosis(df.title_length)
print(
    f"Skewness: {title_length_skewness}, Kurtosis: {title_length_kurtosis}"
)
```

```python
# Print the shortest and longest title based on their length
shortest_title = df.loc[df.title_length.idxmin(), "title"]
longest_title = df.loc[df.title_length.idxmax(), "title"]
print(f"Shortest title: {shortest_title}, Longest title: {longest_title}")
```

```python
# plot your results
plt.hist(df.title_length, bins=20)
plt.title("Distribution of Title Lengths")
plt.xlabel("Title Length")
plt.ylabel("Frequency")
plt.show()
```

## Exercise 2 – Manual Calculation of Standard Deviation

In this exercise, you will manually calculate the standard deviation by selecting five actual titles from the movie dataset you loaded earlier.


```python
# Create a list with 5 actual values from df['title_length'], for example: [10, 13, 14, 18, 22]
actual_values = df["title_length"].head().tolist()
  
```

```python
# Calculate the mean of the 5 values
mean_actual_values = np.mean(actual_values)
print(f"Mean of actual values: {mean_actual_values}")
```

```python
# Calculate the squared differences with respect to the mean
squared_differences = [(x - mean_actual_values) ** 2 for x in actual_values]
print(f"Squared differences: {squared_differences}")
```

```python
variance_actual_values = np.mean(squared_differences)
std_actual_values = np.sqrt(variance_actual_values)
print(f"Variance of actual values: {variance_actual_values}, Standard Deviation of actual values: {std_actual_values}")
```
